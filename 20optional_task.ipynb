```
filepath="/content/car_price_prediction_with_missing.csv"
df=pd.read_csv("car_price_prediction_with_missing.csv")
```

```
df
```

```
print("Independent:")
X=df.drop('Mileage',axis=1)
X.head()
```

```
print("Dependent:")
Y=df['Mileage']
Y.head()
```

```
print("Shape of X:",X.shape)
print("Shape of Y:",Y.shape)
```

```
categorical=X.select_dtypes(include=['object']).columns
print("Categorical columns:",categorical)
X=pd.get_dummies(X,columns=categorical,dtype=int)
X.head()
```

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ["Year", "Engine Size","Price"]

X[num_cols] = scaler.fit_transform(X[num_cols])

print(X.head())
```

```
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(
    X,Y,test_size=0.33,random_state=42)
print("X_train Shape :", X_train.shape)
print("X_test Shape  :", X_test.shape)
print("y_train Shape :", Y_train.shape)
print("y_test Shape  :", Y_test.shape)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
print("model intercept:",model.intercept_)
print("\nmodel coefficients:",model.coef_)
```

```
from numpy._core import numeric
print(X_test.isnull().sum())
X_test=X_test.fillna(X_train.mean(numeric_only=True))
```

```
Y_pred = model.predict(X_test)

print("Predicted Values:")
print(Y_pred[:10])

print("Actual Values:")
print(Y_test.head(10).values)
```

```
import joblib

joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X.columns.tolist(), "columns.pkl")

print("Model Saved Successfully")
```
