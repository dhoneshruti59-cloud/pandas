```
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(2,2,1)
sns.histplot(df["Age"])
plt.title("AGE")
plt.subplot(2,2,2)
sns.histplot(df["Sex"])
plt.title("SEX")
plt.subplot(2,2,3)
sns.histplot(df["Pclass"])
plt.title("PCLASS")
plt.tight_layout()
```

```
sns.countplot(x="Sex",data=df)
plt.title("count of sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
sns.countplot(x="Pclass",data=df)
plt.title("count of pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```
corr_matrix=df[["Age","Pclass"]].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix,annot=True)
plt.title("Heatmap")
plt.show()
```

```
average=df['Age'].mean()
print("average:",average)
max=df['Age'].max()
print("Maximum:",max)
min=df['Age'].min()
print("Minimum:",min)
```

```
plt.figure(figsize=(8,5))
sns.boxplot(x="Fare",y="Pclass",hue="Age",data=df)
plt.title("Boxplot")
plt.xlabel("Fare")
plt.ylabel("Pclass")
plt.show()
```
