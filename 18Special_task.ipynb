```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Missing Values
print(df.isnull().sum())

# Remove Duplicates
df = df.drop_duplicates()

# Histogram
sns.histplot(df['discover_weekly_user'])
plt.show()

# Count Plot
sns.countplot(x=df['subscription'],data=df)
plt.show()

# Heatmap
corr=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f")
plt.show()

# Feature Identification
X = df.drop(columns=["daily_listening_minutes","listener_id"])
y = df["daily_listening_minutes"]

# Encoding
df = pd.get_dummies(
    df,
    columns=[
        "country",
        "platform",
        "subscription",
        "top_genre",
        "top_artist",
        "top_mood"
    ],
    drop_first=True
)

# Scaling
scaler = StandardScaler()

scale_cols = [
    "age",
    "songs_per_day",
    "playlists_count",
    "skip_rate_pct"
]

df[scale_cols] = scaler.fit_transform(df[scale_cols])

print(df.head())
```
